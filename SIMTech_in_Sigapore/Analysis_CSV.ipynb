```python
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
plt.rcParams['font.family'] = "Malgun Gothic"
mpl.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

```python
import pandas
 
def csvread(filename='A.csv'):
    df = pandas.read_csv(filename) # df is pandas.DataFrame
    return df
    
```

```python
# read csv file
df = csvread("E2 TextLog2.txt") # df is pandas.DataFrame
Time = df['Timestamp'].map(lambda x: datetime.strptime(str(x), '%Y/%m/%d %H:%M:%S.%f'))
id1 = df['ns=2;i=201732']
id2 = df['ns=2;i=201972']
id3 = df['ns=2;i=202672']
df.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(Time,id1,label = 'idx1')
plt.plot(Time,id2,label='idx2')
plt.plot(Time,id3,label='idx3')
plt.xlabel('Timestamp')
plt.ylabel('Idx')
plt.ylim(0,15)
plt.title('Idx Graph')
plt.legend()
plt.show()
```
